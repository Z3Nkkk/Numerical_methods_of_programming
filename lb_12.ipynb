> ### 1) Метод Ейлера
> ### а) y'= x + sin(y / sqrt(0.7)); y(1.2) = 1.4; [1.2; 2.2]




```python
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
    return x + np.sin(y / np.sqrt(0.7))

a = 1.2
b = 2.2
h = 0.1
y0 = 1.4

def euler_method(f, a, b, h, y0):
    n = int((b - a) / h) + 1
    x = np.linspace(a, b, n)
    y = np.zeros(n)
    y[0] = y0

    for i in range(1, n):
        y[i] = y[i - 1] + h * f(x[i - 1], y[i - 1])

    return x, y

x, y = euler_method(f, a, b, h, y0)

for i in range(len(x)):
    print(f"x = {x[i]:.2f}, y = {y[i]:.2f}")

plt.plot(x, y, "o", x, y, "black")
plt.title("Метод Ейлера")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

### 2) Ейлера-Коші  
### а) y'= x + sin(y / sqrt(0.7)); y(1.2) = 1.4; [1.2; 2.2]

```python
import numpy as np
import matplotlib.pyplot as plt

def f(y, x):
    return x + np.sin(y / np.sqrt(0.7))

a = 1.2
b = 2.2
h = 0.1
y0 = 1.4

def euler_cauchy_method(f, a, b, h, y0):
    n = int((b - a) / h) + 1
    x = np.linspace(a, b, n)
    y = np.zeros(n)
    y[0] = y0

    for i in range(1, n):
        k1 = h * f(x[i - 1], y[i - 1])
        k2 = h * f(x[i - 1] + h, y[i - 1] + k1)
        y[i] = y[i - 1] + 0.5 * (k1 + k2)

    return x, y

x, y = euler_cauchy_method(f, a, b, h, y0)

for i in range(len(x)):
    print(f"x = {x[i]:.2f}, y = {y[i]:.2f}")

plt.plot(x, y, "o", x, y, "black")
plt.title("Метод Ейлера-Коші")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

### 3) odeint ()  
### а) y'= x + sin(y / sqrt(0.7)); y(1.2) = 1.4; [1.2; 2.2]

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def f(y, x):
    return x + np.sin(y / np.sqrt(0.7))

a = 1.2
b = 2.2
h = 0.1
y0 = 1.4

x_points = np.arange(a, b + h, h)
y_solution = odeint(f, y0, x_points)

for i in range(len(x_points)):
    print(f"x = {x_points[i]:.2f}, y = {y_solution[i][0]:.6f}")

plt.plot(x, y, "o", x, y, "black")
plt.title("odient()")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```
