```python
import numpy as np
from scipy import optimize
from scipy.misc import derivative
import math

x0 = 0.65
y0 = -0.3

def f1(y):
    return 0 - math.cos(y - 2) #задаємо функції

def f2(x):
    return math.sin(x + 0.5) - 1.2 #задаємо функції

def iter (x,y,e):
    xn = x
    yn = y
    xn1 = f1(y)
    yn1 = f2(x)
    n = 1

    while ((abs(xn1 - xn)>=e) & (abs(yn1 - yn) >= e)):
        xn = xn1
        yn = yn1
        xn1 = f1(yn)
        yn1 = f2(xn)
        n += 1

    print ('Метод простої ітерації:')
    print ('x =', xn, '\ny =',yn,'\nКількість ітерацій = ',n)
iter(x0,y0,0.001)

def f3(x):
    return math.sin(x[0] + 0.5) - x[1] - 1.2, math.cos(x[1] - 2) + x[0]

s = optimize.root(f3, [0.,0.], method = 'hybr')

print ('Перевірка',s.x)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

x0 = -2
y0 = -2
x_arr = []
y_arr = []
x = x0
y = y0

while x < 1.5:

  x_arr.append(0 - math.cos(y - 2))
  y_arr.append(math.sin(x + 0.5) - 1.2)
  x += 0.1
  y += 0.1

plt.figure(figsize=(8,7))

axis_y = np.arange(y0, y - 0.1, 0.1)
axis_x = np.arange(x0, x - 0.1, 0.1)

plt.plot(x_arr, axis_x, label = "X")
plt.plot(axis_y, y_arr, label = "Y")

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.grid(True)

plt.title('Відокремлення коренів графічним способом')
plt.xticks(np.arange(x0, x, 0.2))
plt.yticks(np.arange(y0, y, 0.2))

plt.show()
```
